```python
scores=[90,87,66,44,89]
total=sum(scores)
average=total/len(scores)
highest=max(scores)
lowest=min(scores)
print("Total score:", total)
print("Average score:", average)
print("highest score:", highest)
print("lowb score:", lowest)
```

```python
sales = [1000, 2000, 30000, 4000]

total = sum(sales)
avg = total / len(sales)
high_sales = max(sales)
low_sales = min(sales)

print("Total:", total)
print("Average:", avg)
print("high_sales:", high_sales)
print("low_sales:", low_sales)
```
