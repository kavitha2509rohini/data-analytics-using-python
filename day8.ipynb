```python
import matplotlib.pyplot as plt

students = ["A", "B", "C", "D"]
marks = [85, 92, 78, 88]

plt.bar(students, marks, color="skyblue")
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 15, 20]

# Create pie chart
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Programming Language Popularity')
plt.axis('equal')  # Makes the pie chart circular

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Monthly sales data
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [12000, 15000, 18000, 17000, 21000, 25000]
}

df = pd.DataFrame(data)

# Create dashboard
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(df["Month"], df["Sales"], marker="o", linewidth=2)
ax.set_title("Monthly Sales Dashboard")
ax.set_xlabel("Month")
ax.set_ylabel("Sales ($)")
ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample sales data
sales = [120, 150, 180, 170, 210, 250, 190, 220, 240, 160, 175, 200]

# Create histogram
plt.hist(sales, bins=5, color='skyblue', edgecolor='black')

plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 65, 70, 75, 85, 90, 95]

# Create scatter plot
plt.scatter(study_hours, exam_scores)

plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr"]
sales = [100, 150, 200, 180]
profit = [20, 35, 50, 45]

# Create 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# First subplot
axes[0].plot(months, sales, marker='o')
axes[0].set_title("Sales")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Sales")

# Second subplot
axes[1].bar(months, profit)
axes[1].set_title("Profit")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Profit")

plt.tight_layout()
plt.show()
```
