```python
df=df.drop_duplicates()
print(df)
```

```python
total_Calories = df["Calories"].sum()
print("total_Calories:", total_Calories)

Duration = df.groupby("Maxpulse")["Calories"]
print(Duration)
```

```python
Duration.plot(kind='bar')
plt.title("pulse maxpulse")
plt.xlabel("Pulse")
plt.ylabel("Maxpulse")
plt.show()
```
