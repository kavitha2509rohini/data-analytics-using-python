```python
marks = {
    "Tamil": 85,
    "English": 78,
    "Maths": 92,
    "Science": 88,
    "Social": 81
}

total = 0

for subject, mark in marks.items():
    print(subject, ":", mark)
    total += mark

print("Total Marks =", total)

if total >= 450:
    print("Grade A")
elif total >= 400:
    print("Grade B")
elif total >= 350:
    print("Grade C")
else:
    print("Grade D")
```
